Cálculo de la tendencia de crímen Tendencia de crimen

$\text{delitos relativos año presente} = \frac{\text{Delitos año presente}}{\text{Población año presente}}$

$\text{delitos relativos año pasado} = \frac{\text{Delitos año pasado}}{\text{Población año pasado}}$

$\text{Tendencia de crímen} = \frac{\text{Delitos relativos año presente}-\text{delitos relativos año pasado}}{\text{delitos relativos año pasado}}$

```python
df_delitos_relativo = df_delito/(df_poblacion_2015_2024) #delitos cometidos respecto al total de la población en ese estado
df_delitos_relativo_p = df_delito_p/(df_poblacion_2015_2024) #delitos (enero-mayo) cometidos respecto al total de la población en ese estado
df_tendencia_crimen = pd.DataFrame()

for año in range(2016, 2025):
    if año==2024:
      df_tendencia_crimen[str(año)] = ((df_delitos_relativo_p[str(año)]-df_delitos_relativo_p[str(año-1)])/df_delitos_relativo_p[str(año-1)]).replace(np.nan, 0).replace(np.inf, 0)
    else:
      df_tendencia_crimen[str(año)] = ((df_delitos_relativo[str(año)]-df_delitos_relativo[str(año-1)])/df_delitos_relativo[str(año-1)]).replace(np.nan, 0).replace(np.inf, 0)

df_tendencia_crimen = df_tendencia_crimen*100
df_tendencia_crimen.reset_index(inplace=True)
df_tendencia_crimen.rename(columns={'index':'Estado'}, inplace=True)
df_tendencia_crimen
```
